```python
# Extract population data for selected countries (1960–2023)
countries = ["United States", "India", "China", "Russian Federation"]
population_data = data.loc["1960":"2023", countries]

# Reshape data for Seaborn
population_melted = population_data.reset_index().melt(
    id_vars="index",
    value_vars=countries,
    var_name="Country",
    value_name="Population"
)

# Rename and clean
population_melted.rename(columns={"index": "Year"}, inplace=True)
population_melted["Year"] = population_melted["Year"].astype(int)

# Set Seaborn theme
sns.set(style="whitegrid")

# Custom color palette for each country
country_colors = {
    "United States": "skyblue",
    "India": "green",
    "China": "red",
    "Russian Federation": "purple"
}

# Create FacetGrid with larger canvas (wider and taller)
g = sns.FacetGrid(
    population_melted,
    col="Country",
    col_wrap=2,
    height=5,          # Height of each subplot
    aspect=1.5,        # Width-to-height ratio of each subplot
    sharex=True,
    sharey=False
)

# Custom plotting function
def plot_country(data, color=None, **kwargs):
    country = data["Country"].iloc[0]
    sns.lineplot(
        x="Year",
        y="Population",
        data=data,
        color=country_colors[country],
        marker="o",
        linewidth=2
    )

# Map plotting function
g.map_dataframe(plot_country)

# Set subplot titles and axis formatting
g.set_titles("{col_name}")
for ax in g.axes.flat:
    ax.set_xticks(range(1960, 2025, 5))  # Show year every 10 years
    ax.set_xticklabels(range(1960, 2025, 5), rotation=45)
    ax.grid(True, alpha=0.3)

# Add global X and Y labels
g.set_axis_labels("", "")
g.fig.text(0.5, 0.04, 'Year', ha='center', va='center', fontsize=14)
g.fig.text(0.04, 0.5, 'Population', ha='center', va='center', rotation='vertical', fontsize=14)

# Add overall title
g.fig.suptitle("Population Growth (1960–2023)", fontsize=18, y=1.03)

# Adjust layout to avoid overlap
plt.tight_layout(rect=[0.05, 0.05, 1, 0.97])

# Show the plot
plt.show()
```

```python
# Choose countries to compare
countries = ["China", "India", "United States", "Brazil", "Russian Federation"]

# Prepare data for Seaborn (reshape to long format)
data_melted = data[countries].reset_index().melt(
    id_vars="index", var_name="Country", value_name="Population"
)
data_melted.rename(columns={"index": "Year"}, inplace=True)
data_melted["Year"] = data_melted["Year"].astype(int)

# Set Seaborn style
sns.set(style="whitegrid")

# Create the line plot
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=data_melted,
    x="Year",
    y="Population",
    hue="Country",
    palette="cubehelix",
    marker="o"
)

# Set title and labels
plt.title("Population Growth Comparison (1960–2023)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Population")

# Set x-axis ticks every 5 years
plt.xticks(np.arange(1960, 2025, 5))

# Make gridlines lighter
plt.grid(True, alpha=0.3)   # alpha controls transparency (0 = invisible, 1 = solid)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python

# Select specific countries for visualization
selected_countries = ["China", "India", "United States", "Brazil", "Indonesia"]

# Transpose data so countries appear on the y-axis and years on the x-axis
heatmap_data = data[selected_countries].T

# Set Seaborn style for clean visuals
sns.set(style="white")

# Create a figure with custom size
plt.figure(figsize=(10, 6))

# Draw the heatmap
sns.heatmap(
    heatmap_data,
    cmap="cubehelix",       # Use a Cubehelix color map (visually appealing for population data)
    annot=False,          # Set to True to show actual values in cells
    cbar_kws={'label': 'Population (in billions)'},  # Add colorbar label
)

# Add descriptive title and axis labels
plt.title("Population Heatmap (1960–2023)", fontsize=14)
plt.ylabel("Country")
plt.xlabel("Year")

# Adjust layout to prevent clipping
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Calculate year-over-year population growth rate (%) for China
growth_rate = data["China"].pct_change() * 100

# Apply Seaborn style for better aesthetics
sns.set(style="whitegrid")

# Create figure with defined size
plt.figure(figsize=(10, 8))

# Plot line chart using Seaborn
sns.lineplot(
    x=data.index.astype(int),   # Convert index to integer (year)
    y=growth_rate,              # Growth rate values
    color="red",                # Line color
    marker="o"                  # Show data points
)

# Add chart title and axis labels
plt.title("China's Population Growth Rate (1960–2023)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Growth Rate (%)")

# Set x-axis ticks every 5 years
plt.xticks(np.arange(1960, 2025, 5))

# Adjust grid transparency for subtle effect
plt.grid(True, alpha=0.3)

# Prevent element overlap
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Apply Seaborn's clean white grid style for better visuals
sns.set(style="whitegrid")

# Create a figure with defined size
plt.figure(figsize=(10, 6))

# Plot global population trend using Seaborn
sns.lineplot(
    x=data.index.astype(int),   # Convert index to integer (year values)
    y=data["World"],            # Global population data
    color="skyblue",               # Line color
    marker="o"                  # Add circular markers for each data point
)

# Add descriptive title and axis labels
plt.title("Global Population Growth (1960–2023)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Population")

# Display ticks on the x-axis every 5 years
plt.xticks(np.arange(1960, 2025, 5))


# Add gridlines for readability (default whitegrid already adds them)
plt.grid(True, alpha=0.3)  # Adjust opacity for a softer grid look

# Prevent overlapping of plot elements
plt.tight_layout()

# Display the final chart
plt.show()
```
